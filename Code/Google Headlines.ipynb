```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from lxml.html import fromstring
#Load Broadway Original
Broadway_Original = pd.read_csv('/Users/morganhogenmiller/Desktop/DA 401/Data sets/broadway(corgi) copy.csv') #Read in CORGI data set

#Load shows in Broadway average (defined by the shows used in Wikipedia synopses)
os.chdir ('/Users/morganhogenmiller/Desktop/DA 401/Data sets')
Broadway_Avg = pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Final_Musicals.csv') #Read in CORGI data set

#Provides a list of shows in form needed for function 
show_list=list(Broadway_Avg['Show']) 
```

```python
from textblob import TextBlob #Allows us to analyze subjectivity
from bs4 import BeautifulSoup
import requests

class Analysis: 
    def __init__(self, show):
        """
        This function initializes all of the objects used for gathering Google headlines and 
        sentiment analysis. It takes the parameters self and show. Self corresponds to the Google page of interest, and show is the search show whose headlines are going to be analyzed. 
        """
        self.show= show
        self.sentiment=0
        self.url='https://www.google.com/search?q={}&source=lnms&tbm=nws'.format(self.show) #Define search page
    def run(self):
        """
        The run function gathers the html document and the headlines for the ur for the search show identified and uses beautiful soup to 
        print the full html. It takes self (the url) as a parameter.
        """
        response=requests.get(self.url) #sends request to specific url and gets raw html
        soup=BeautifulSoup(response.text, 'html.parser')
        headline_results=soup.find_all('div', attrs={'class':'g'})
        for h in headline_results:
            title=(h.find('h3')).get_text()
            title=TextBlob(title)
            self.sentiment += title.sentiment.polarity / len(headline_results) #polarity between (0-1) of the 
        print(self.sentiment)
        
a=Analysis('The Color Purple')
a.run()

#print(a.show, 'Sentiment:',a.sentiment)

# Advanced search url
#https://www.google.com/search?hl=en&as_q=Hamilton+musical&as_epq=&as_oq=&as_eq=&as_nlo=&as_nhi=&lr=&cr=&as_qdr=y&as_sitesearch=&as_occt=any&safe=images&as_filetype=&as_rights=

# Regular search url
#https://www.google.com/search?q=Hamilton+musical&lr=&hl=en&tbas=0&source=lnt&sa=X&ved=0ahUKEwiRkvDarf3gAhVyjoMKHWDADGwQpwUIIw&biw=622&bih=721
```

```python
#Making a dataframe with the sentiment scores of all of the shows
show_sent=[]
for i in show_list:
        a=Analysis(i)
        a.run()
        show_sent.append(a.sentiment)
        
show_sentimentality=pd.DataFrame({'Name':show_list ,'Sentimentality':show_sent})     
show_sentimentality.sort_values(by=['Name'])
```

```python
sents=list(show_sentimentality['Sentimentality'])
show_sentimentality
```

```python
#Merge show sentimentality with weeks
#Isolating relevant shows and thier longevity
shows137=pd.DataFrame(Broadway_Original[Broadway_Original['Name'].isin(show_list)]) #Cleaning Broadway_Original data for only 137 shows 
weeks=pd.DataFrame(shows137['Name'].value_counts())
weeks=weeks.reset_index()
weeks=weeks.rename(index=str, columns={"index": "Name", "Name": "Weeks"})
weeks=weeks.sort_values(by='Name')
```

```python
#Merging number of weeks with the sentimentality of shows
sents=weeks.join(show_sentimentality.set_index('Name'), on='Name')

#Export to csv
pd.DataFrame.to_csv(sents,'google_headlines.csv')
sents
```
