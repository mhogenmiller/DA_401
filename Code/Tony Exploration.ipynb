```python
#Creating data set with Boolean variables for if a show won a Tony in each major category

def tonys(shows):
    """
    This function takes a list of shows and creates a boolean variable for each award category in the big four, 
    ultimately to get information for if the show won a Tony in each of the categories. The function takes "shows"
    as a parameter, which is a list of the shows to be examined in relation to the Tony data set from years 2008 to 2016.
    """

    musical=[]
    revival=[]
    director=[]
    actor=[]
    actress=[]

    for i in show_list:
        if i in list(tony.iloc[:,1]): #add 1 for best musical winners
            musical.append(1)
        else: 
            musical.append(0)

        if i in list(tony.iloc[:,2]): #add 1 for best revival
            revival.append(1)
        else: 
            revival.append(0)

        if i in list(tony.iloc[:,3]): #add 1 for best director
            director.append(1)
        else: 
            director.append(0)
        if i in list(tony.iloc[:,5]): #add 1 for best actor
            actor.append(1)
        else: 
            actor.append(0)
        if i in list(tony.iloc[:,4]): #add 1 for best actress
            actress.append(1)
        else: 
            actress.append(0)


    Tony=pd.DataFrame({'Name': show_list, 'Best Musical':musical, 'Best Revival':revival, 'Best Director':director, 'Best Actor':actor,'Best Actress':actress})
    return Tony
```

```python
Tony=tonys(show_list)
```

```python
#Adding a column to the data for the total number of Tonys each show wins
Tony['total']=Tony['Best Musical']+Tony['Best Revival']+Tony['Best Director'] + Tony['Best Actor'] + Tony['Best Actress']
weeks= weeks.loc[:,['Weeks','Name']]

#Merging Tony data with number of weeks a show was on Broadway 
Tony=Tony.join(weeks.set_index('Name'),on='Name')
Tony
```
