```python
#going to try removing the negative and 0 values to see if that will improve the accuracy of the model
df=df[df.Category != 0]
df=df[df.Category != 'Negative']
```

### Pre-Processing ###
#Should I remove all negatives so that I am not working with such an uneven distribution?

```python
#From synopses to list of words in the synopsis
from nltk.tokenize import word_tokenize
words=[]
for syn in list(df['Synopsis']):
    tokenized_word=word_tokenize(syn)
    words.append(tokenized_word)
```

```python
#making lowercase and moving stop words
from nltk.corpus import stopwords
stop_words=list(stopwords.words("english"))
#filtered_sent=[]
allwords=[]

for w in words:
    filtered_sent=[]
    for word in w:
        word=word.lower() #Make all lowercase
        if word not in stop_words:
            filtered_sent.append(word)
    allwords.append(filtered_sent)

allwords[0]
#Lematization?[1]
#Don't think I really NEED to do anything else?
```

```python
#merge lists into sentences (or what's left of them without stop words)
df2=[]
for s in allwords:
    s=(" ".join(s))
    df2.append(s)
```

```python
#Make new columns without stop words
nostops=pd.DataFrame({'Sans SW':df2})
nostops.reset_index()
```

### Feature Generation using TF-IDF ###

IDF(Inverse Document Frequency) measures the amount of information a given word provides across the document. IDF is the logarithmically scaled inverse ratio of the number of documents that contain the word and the total number of documents.

```python
#Creating a document term matrix to see what words are in which documents
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
#tokenizer to remove unwanted elements from out data like symbols and numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(tokenizer = token.tokenize)
text_counts= cv.fit_transform(nostops['Sans SW'])
print(text_counts.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
text_tf= tf.fit_transform(nostops['Sans SW'])
```

```python
#Split into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_tf, nostops['Sans SW'], test_size=0.3, random_state=50)
```

```python
#Building and training model and testing on the test set
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
# Model Generation Using Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```
