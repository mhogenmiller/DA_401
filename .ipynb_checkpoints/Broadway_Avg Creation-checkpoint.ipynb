```python
import pandas as pd
import os

#Filter Broadway_Orignial 
Broadway_Original = pd.read_csv('/Users/morganhogenmiller/Desktop/DA 401/Data sets/broadway(corgi) copy.csv') #Read in CORGI data set

#Filter to shows after 2007 only
Broadway_Original=Broadway_Original.loc[Broadway_Original['Year']>2007] 
Broadway_Original=Broadway_Original.loc[Broadway_Original['Type']=='Musical']
Broadway_Avg = pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Final_Musicals.csv') #Read in CORGI data set

#Provides a list of shows in form needed for function 
show_list=list(Broadway_Avg['Show'])
Broadway_O=pd.DataFrame(Broadway_Original[Broadway_Original['Name'].isin(show_list)])

#Average attendence, capacity, Day, Gross, Gross Potential,
Broadway_Original=Broadway_O.iloc[:,0:12]

#Importing Google Headlines, Wikipedia, Google Trends, and Tonys data
gh=pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Data sets/google_headlines.csv')
w=pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Data sets/Self_Scores_137.csv')
gt=pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Data sets/GT.csv')
t=pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Tonys.csv')
censored=pd.read_csv('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Data sets/Censored.csv')
t=t.drop(['Unnamed: 0'],axis=1)
t=t.drop(['Weeks'], axis=1)
gt=gt.drop(['Unnamed: 0'],axis=1)
w=w.drop(['Unnamed: 0'],axis=1)
w=w.rename(columns={'Category':'Category','Show':'Name','Synopsis':'Synopsis'})

#Joining all data sets together
ghw=gh.join(w.set_index('Name'), on='Name')
ghwt=ghw.join(gt.set_index("Name"),on='Name')
ghwtt=ghwt.join(t.set_index("Name"),on='Name')
ghwtt=ghwtt.drop(['Unnamed: 0'],axis=1)
ghwtt=ghwtt.join(censored.set_index('Name'),on='Name')
```

```python
#Averaging original Broadway data (attendence, gross, etc.) for each show
Broadway_Original=Broadway_Original.iloc[:,0:9]
Broadway_Original=Broadway_Original.groupby(['Name']).mean()
Broadway_Original=Broadway_Original.reset_index()
Broadway_Original=Broadway_Original.rename(columns = {'Show':'Name'})
```

```python
#Combining new and old variables aggregated into Broadway_Average
#Make shows names in the Broadway statistics
Broadway_Avg=Broadway_Original.join(ghwtt.set_index("Name"),on='Name')
os.chdir('/Users/morganhogenmiller/Desktop/GitHub/DA_401/Data sets')
pd.DataFrame.to_csv(Broadway_Avg, 'Broadway_Avg.csv')
Broadway_Avg
```
